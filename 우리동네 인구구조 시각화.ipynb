```python
# 성별 데이터 저장하기 (Type 1)

import csv

f = open('gender_new.csv', encoding='cp949')
data = csv.reader(f)
m = []
f = []

for row in data:
    if '신도림' in row[0]:
        for i in range(0, 101):
            m.append(int(row[i+3]))  # 3~103번 인덱스 데이터 순차적으로 저장
            f.append(int(row[-(i+1)]))   # -1부터 -101 인덱스 데이터 순서대로 저장한 후
            
f.reverse()   # 뒤집기
```

```python
# 성별 데이터 저장하기 (Type 2)

f = open('gender_new.csv', encoding='cp949')
data = csv.reader(f)
m = []
f = []

for row in data:
    if '신도림' in row[0]:
        for i in row[3:104] :    # 3 ~ 103번 데이터 슬라이싱   
            m.append(int(i))
            #m.append(-int(i))     # 정수 데이터로 변환, m 리스트에 저장
        for i in row[106:] :   # 106 ~ 끝까지 데이터 슬라이싱
            f.append(int(i))   # 정수 데이터로 변환, f 리스트에 저장
```

```python
# 성별 데이터 수평 막대그래프로 시각화하기

import matplotlib.pyplot as plt

plt.barh(range(101), m, label="male")
plt.barh(range(101), f, label="female")
plt.legend()
plt.show()
```

```python
# 성별 데이터 수평 막대 그래프로 시각화 하기
# 남성 데이터 음수로 바꾸기
# m.append(-int(i))

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title("신도림 지역의 남녀 성별 인구 분포")
plt.barh(range(101), m, label="male")
plt.barh(range(101), f, label="female")
plt.legend()
plt.show()
```

```python
# 찾고 싶은 지역의 이름을 입력해서 항아리 모양 그래프 그리기

f = open('gender_new.csv', encoding='cp949')
data = csv.reader(f)
m = []
f = []

name = input('찾고싶은 지역의 이름을 알려주세요 : ')
for row in data:
    if name in row[0]:
        for i in row[3:104] :    # 3 ~ 103번 데이터 슬라이싱   
            m.append(-int(i))     # 정수 데이터로 변환, m 리스트에 저장
        for i in row[106:] :   # 106 ~ 끝까지 데이터 슬라이싱
            f.append(int(i))   # 정수 데이터로 변환, f 리스트에 저장
            
plt.style.use('ggplot')
plt.figure(figsize=(10, 5), dpi=300)
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name+' 지역의 남여 성별 인구 분포')
plt.barh(range(101), m, label='남성')
plt.barh(range(101), f, label='여성')
plt.legend()
plt.show()
```

# Unit 9 우리 동네 인구 구조를 파이 차트로 나타내기

```python
# 제주특별자치도의 경우, 제주특별자치도 뿐만아니라 해당 도의 여러 읍들이 함께 저장되어있음
# 따라서, 첫번째 데이터만 저장하고 반복문을 빠져나갈 것!

# 찾고 싶은 지역의 이름을 입력해서 항아리 모양 그래프 그리기

f = open('gender_new.csv', encoding='cp949')
data = csv.reader(f)
m = []
f = []

name = input("찾고 싶은 지역의 이름을 알려주세요: ")

for row in data:
    if name in row[0]:
        for i in row[3:104] :    # 3 ~ 103번 데이터 슬라이싱   
            m.append(-int(i.replace(',', '')))     # 정수 데이터로 변환, 쉼표 제거 후 m 리스트에 저장
        for i in row[106:] :   # 106 ~ 끝까지 데이터 슬라이싱
            f.append(int(i.replace(',', '')))   # 정수 데이터로 변환, 쉼표 제거 후 f 리스트에 저장
            
        break   # 첫번째 데이터만 저장하고 반복문 빠져나가기!
            
plt.style.use('ggplot')
plt.figure(figsize=(10, 5), dpi=300)
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name+' 지역의 남여 성별 인구 분포')
plt.barh(range(101), m, label='남성')
plt.barh(range(101), f, label='여성')
plt.legend()
plt.show()
```

```python
# 성별 연령별 인구 데이터 출력 및 실행

import csv

f = open('gender_new.csv', encoding='cp949')
data = csv.reader(f)
size = []

name = input("찾고 싶은 지역의 이름을 알려주세요 : ")
for row in data:
    if name in row[0]:
        m = 0
        f = 0
        for i in range(101):
            m += int(row[i+3].replace(',', ''))    # Remove comma and convert to integer
            f += int(row[i+106].replace(',', ''))  # Remove comma and convert to integer
        break

size.append(m)    # 남성 합계 데이터 리스트에 추가
size.append(f)    # 여성 합계 데이터 리스트에 추가
print(size)
```

```python
# 파이차트로 표현하기

f = open('gender_new.csv', encoding='cp949')
data = csv.reader(f)
size = []

name = input("찾고 싶은 지역의 이름을 알려주세요 : ")
for row in data:
    if name in row[0]:
        m = 0
        f = 0
        for i in range(101):
            m += int(row[i+3].replace(',', ''))    # Remove comma and convert to integer
            f += int(row[i+106].replace(',', ''))  # Remove comma and convert to integer
        break

size.append(m)    # 남성 합계 데이터 리스트에 추가
size.append(f)    # 여성 합계 데이터 리스트에 추가

plt.rc('font', family='Malgun Gothic')
color = ['crimson', 'darkcyan']
plt.axis('equal')
plt.pie(size, labels=['남', '여'], autopct="%.1f%%", colors=color, startangle=90)
plt.title(name + ' 지역의 남녀 성별 비율')
plt.show()
```

## 꺾은선 그래프로 표현

```python
# 성별 연령별 인구 데이터 저장하기
f = open('gender_new.csv', encoding='cp949')
data = csv.reader(f)

m = []
f = []

name = input("궁금한 동네를 입력 해주세요 : ")
for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i].replace(',', '')))    # Remove comma and convert to integer
            f.append(int(row[i+103].replace(',', '')))  # Remove comma and convert to integer
        break
    
# 성별 연령별 인구 데이터 꺾은선 그래프로 표현하기
plt.plot(m, label='Male')
plt.plot(f, label='Famale')
plt.legend()
plt.show()
```

## 막대 그래프로 표현

```python
# 연령별 남성 인구 - 여성 인구 값 저장하기

f = open('gender_new.csv', encoding='cp949')
data = csv.reader(f)

result = []

name = input("궁금한 동네를 입력 해주세요 : ")
for row in data:
    if name in row[0]:
        for i in range(3,104):
            result.append(int(row[i].replace(',', '')) - int(row[i+103].replace(',', '')))    # Remove comma and convert to integer
        break
    
import matplotlib.pyplot as plt
plt.bar(range(101), result)
plt.show()
```

### 제주도의 연령대별 성별 비율

```python
import csv
f = open('gender_new.csv', encoding='cp949')
data = csv.reader(f)

m = []
f = []

name = input('궁금한 동네를 입력해주세요 : ')
for row in data:
    if name in row[0]:
        for i in range(3, 104):
            m.append(int(row[i].replace(',', '')))
            f.append(int(row[i+103].replace(',', '')))
        break
    
plt.style.use('ggplot')
plt.scatter(m, f)
plt.show()
```

```python
# 추세선, 원의 크기 추가
import csv
import math

f = open('gender_new.csv', encoding='cp949')
data = csv.reader(f)

m = []
f = []
size = []

name = input('궁금한 동네를 입력해주세요 : ')
for row in data:
    if name in row[0]:
        for i in range(3, 104):
            m.append(int(row[i].replace(',', '')))
            f.append(int(row[i+103].replace(',', '')))
            size.append(math.sqrt(int(row[i].replace(',', '')) + int(row[i+103].replace(',', ''))))
        break
    
plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(10, 5), dpi=300)
plt.title(name+' 지역의 성별 인구 그래프')

plt.scatter(m, f, s=size, c=range(101), alpha=0.5, cmap='jet')
plt.colorbar()

plt.plot(range(max(m)), range(max(m)), 'g')

plt.xlabel('남성 인구 수')
plt.ylabel('여성 인구 수')
plt.show()
```

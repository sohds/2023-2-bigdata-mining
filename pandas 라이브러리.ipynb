#### 금메달 기준으로 데이터 정렬하기(내림차순)

```python
import pandas as pd
df = pd.read_html('https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table', header = 0, index_col = 0) 
summer = df[1].iloc[:,:5]
summer.columns = ['경기수', '금', '은', '동', '계']

# sort_values 정렬, ascending false = 오름차순 false, 즉 내림차순 진행!
print(summer.sort_values('금', ascending = False, inplace=True))  # inplace = True로 지정해야 그 모습으로 변함!
```

```python
summer.to_excel('하계올림픽메달.xlsx')
```

## 인구 구조 분석하기

```python
import pandas as pd

# 데이터 읽어오기
df = pd.read_csv('age_new.csv', encoding='cp949', index_col = 0)
df.head()
```

```python
df.info()
```

```python
# df.dtypes

# 콤마를 제거하고 정수로 변환하는 함수 정의
def remove_comma_and_convert_to_int(column):
    return column.str.replace(',', '').astype(int)

# 콤마가 포함된 열에 대해 함수 적용
for col in df.columns:
    if df[col].dtype == 'object':
        try:
            df[col] = remove_comma_and_convert_to_int(df[col])
        except ValueError:
            # 형변환이 실패하면 해당 열은 건너뜀
            continue

df.info()
```

```python
df.head()
```

```python
df = df.div(df['2023년10월_계_총인구수'], axis =0)     # 전체 데이터를 총인구수로 나눠서 비율로 변환
del df['2023년10월_계_총인구수'], df['2023년10월_계_연령구간인구수']  # 총인구수, 연령구간인구수 열 삭제
```

```python
# 열 이름에서 '2023년10월_계' 부분을 제거
df = df.rename(columns=lambda x: x.replace('2023년10월_계_', ''))

# 변경된 열 이름 확인
df.head()
```

```python
name = input('원하는 지역의 이름을 입력해주세요 : ') #2. 지역 이름 입력
a = df.index.str.contains(name) #3. 해당 행을 찾아서 해당 지역의 인구 구조를 저장
df2 = df[a]
df2
```

# Unit 13 숫자 데이터를 쉽게 다루게 돕는 numpy 라이브러리

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0., 5., 0.2)    # np.arange(시작, 끝, 증가분)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
print(np.sqrt(2))  # 제곱근 출력, 2의 제곱근
```

```python
# 파이와 삼각함수 활용하기

print(np.pi)
print(np.sin(0))
print(np.cos(np.pi))
```

```python
# random 서브 라이브러리의 rand 함수 / numpy의 ndarray 타입
# ndarray: n차월 배열 이라는 뜻!

a = np.random.rand(5)    # 0 ~ 1 사이의 임의의 값을 random으로 5개
print(a)
print(type(a))

print(np.random.choice(6, 10))   # (첫번째 칸) 6: 0 ~ 5까지의 숫자 중 고르겠다, 이 자리엔 리스트가 와도 OK! / (두번째 칸) 10: 10개 pick
print(np.random.choice(10, 6, replace=False))   # replace = False: 중복을 허용하지 않는다
print(np.random.choice(6, 10, p=[0.1, 0.2, 0.3, 0.2, 0.1, 0.1]))    # 확률 설정 가능
```

```python
dice = np.random.choice(6, 1000000, p=[0.1, 0.2, 0.3, 0.2, 0.1, 0.1])
plt.hist(dice, bins=6)       # 0, 1, 2, 3, 4, 5 중 랜덤으로 추출한 숫자를 히스토그램으로 표현
plt.show()
```

```python
x = np.random.randint(10, 100, 200)
y = np.random.randint(10, 100, 200)
size = np.random.rand(200) * 100    # 강의교재에는 100으로 되어있는데, x도 y도 전부 200개씩 뽑았기 때문에 200으로 넣어야 함

plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```

* numpy array
  * 행렬 연산이 가능해진다 (vs. 리스트: 얘는 그냥 데이터를 담는 container여서 comma로 split 되어있음)

```python
a = np.array([1, 2, 3, 4])
print(a)
```

```python
# numpy array의 인덱싱과 슬라이싱

print(a[1], a[-1])    # a의 1번 인덱스 값, -1번 인덱스 값 출력
print(a[1:])          # a의 1번 인덱스를 기준으로 슬라이싱 결과 출력
```

```python
# zeros() 함수로 numpy array 만들기

a = np.zeros(10)    # 0으로 이루어진 크기가 10인 배열 생성
print(a)
```

```python
# 연속된 숫자의 numpy array 만들기

print(np.arange(3))  # arange() 함수에 1개 값 입력: 해당 값 미만의 수를 원소로 가진 배열 생성
print(np.arange(3, 7))   # arange() 함수에 2개 값 입력: 처음 값을 포함하고, 마지막 값 미만의 수를 원소로 가진 배열 생성
print(np.arange(3, 7, 2))   # arange() 함수에 3개 값 입력: range()함수 처럼, 마지막 값은 차라고 생각하면 됨
```

```python
# 연속된 실수의 numpy array 만들기

a = np.arange(1, 2, 0.1)    # 1이상 2미만 구간에서 0.1 간격으로 실수 생성
b = np.linspace(1, 2, 11)   # 1부터 2까지 11개 구간으로 나눈 실수 생성
print(a)
print(b)
```

```python
a = np.arange(-np.pi, np.pi, np.pi/10)
b = np.linspace(-np.pi, np.pi, 20)
print(a)
print(b)

# 빼서 사용하는 것과 구간을 나누는 것은 조금 다르게 결과가 나온다.
```

```python
# numpy array에 값을 한꺼번에 더하기

a = np.zeros(10) + 5
print(a)
```

```python
# numpy array에 함수 적용하기
# np.array에 함수 적용하면 모든 요소가 그 함수 적용된 값으로 바뀜!

a = np.linspace(1, 2, 11)
print(a)
print(np.sqrt(a))
```

```python
# numpy를 활용한 그래프 그리기

a = np.arange(-np.pi, np.pi, np.pi/100)
plt.plot(a, np.sin(a))
plt.show()
```

```python
# numpy를 활용한 그래프 그리기

a = np.arange(-np.pi, np.pi, np.pi/100)
plt.plot(a, np.sin(a))
plt.plot(a, np.cos(a))
plt.plot(a+np.pi/2, np.sin(a))
plt.show()
```

```python
# mask 적용하기 (기본 데이터 만들기)

a = np.arange(-5, 5)
print(a)
print(a<0)   # mask 만들기
print(a[a<0])   # mask 적용하기

mask1 = abs(a) > 3
print(a[mask1])

mask2 = abs(a) % 2 == 0
print(a[mask1+mask2])    # 둘 중 하나의 조건이라도 참일 경우 (or)
print(a[mask1*mask2])    # 두 가지 조건이 모두 참일 경우 (and)
```

```python
x = np.random.randint(-100, 100, 1000)  # 1000개의 랜덤 값 추출
y = np.random.randint(-100, 100, 1000)  # 1000개의 랜덤 값 추출
size = np.random.rand(1000) * 100  # Ensure size has the same length as x and y

mask1 = abs(x) > 50
mask2 = abs(y) > 50

x = x[mask1+mask2]
y = y[mask1+mask2]     # mask1과 mask2 중 하나라도 만족하는 값 저장

size = size[:len(x)]   # Truncate the size array to match the size of x and y

plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```

```python
## 궁금한 지역의 인구 구조와 가장 비슷한 곳의 인구 구조를 시각화

f = open('age_new.csv', encoding='cp949')
data = csv.reader(f)
next(data)
data = list(data)

name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요 : ')
mn = 1
result_name = ''
result = 0

for row in data:
    if name in row[0]:
        home = np.array(row[3:], dtype=int)/int(row[2])
        break
    
for row in data:
    away = np.array(row[3:], dtype=int)/int(row[2])
    s = np.sum((home-away)**2)
    if s < mn and name not in row[0]:
        mn = s
        result_name = row[0]
        result = away
        

plt.style.use('ggplot')
plt.figure(figsize=(10, 5), dpi=300)
plt.rc('font', family="Malgun Gothic")
plt.title(name+' 지역과 가장 비슷한 인구 구조를 가진 지역')
plt.plot(home, label=name)
plt.plot(away, label=result_name)
plt.legend()
plt.show()
```
